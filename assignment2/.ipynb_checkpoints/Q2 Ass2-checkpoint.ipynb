```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Question 2(a)

```python
df=pd.read_csv('rs_1.csv',names=['s'])
```

```python
df.head()
```

```python
samples=np.array(df['s'])
used=samples[samples!=-1]
print("{:,} number of samples were rejected out of 100,000 samples".format(100000-used.shape[0]))
```

```python
r_T=[]
r_F=[]
N=[]
sT=[]
sF=[]
for i,n in enumerate(used):
    if n==1:
        r_T.append(1)
    elif(n==2):
        r_F.append(1)
        
    N.append(i+1)
    sT.append(sum(r_T)/(i+1))
    sF.append(sum(r_F)/(i+1))
```

Alternate way of plotting with all samples on x-axis. 

```python
r_T=[]
N=[]
prob=[]
c=0
for i,n in enumerate(samples):
    if n!=-1:
        c+=1
        if n==1:
            r_T.append(n)
        N.append(c)
    if c==0:
        prob.append(0)
        continue
    prob.append(sum(r_T)/(c))
plt.plot((np.arange(1,100001,1)),prob)
```

```python


plt.figure(figsize=(18,10))
plt.yticks(np.arange(0, 1, 0.01))
plt.xticks(np.arange(0, 15, 1))
plt.grid(color='r', linestyle='--', linewidth=2,alpha=0.1)
plt.xlabel('Log of number of accepted samples',size=20)
plt.ylabel('Probability of R=True',size=20)

plt.plot(np.log(N),sT,label='P(R=T|s,w)')


plt.title('P(R=T|s,w) by rejection samples')
plt.savefig('Q2a')
```

# Question 2(b)

```python
print('Out of 100,000 samples there were {:,} samples accepted'.format(used.shape[0]))
```

```python
r_T=[]
N=[]
s=[]
up_epsilon=[]
down_epsilon=[]
for i,n in enumerate(used):
    if n==1:
        r_T.append(n)
    N.append(i+1)
    s.append(sum(r_T)/(i+1))
    epsilon=np.sqrt(1.844/(i+1))
    up_epsilon.append((sum(r_T)/(i+1))+epsilon)
    down_epsilon.append((sum(r_T)/(i+1))-epsilon)
```

```python
plt.figure(figsize=(18,10))
plt.yticks(np.arange(-2, 2, 0.1))
plt.xticks(np.arange(0, 15, 1))
plt.grid(color='r', linestyle='--', linewidth=2,alpha=0.1)
plt.xlabel('Log of number of accepted samples',size=20)
plt.ylabel('Probability of R=True',size=20)
plt.plot(np.log(N),s,label='$s$')
plt.plot(np.log(N),up_epsilon,linestyle='dashed',label='$s+\epsilon$')
plt.plot(np.log(N),down_epsilon,linestyle='dashed',label='$s-\epsilon$')
plt.legend(fontsize='x-large')
plt.savefig('Q2b_2')
```
